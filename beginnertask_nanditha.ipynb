```python
#Matplotlib
#lineplot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
#scatterplot
x = [5, 7, 8, 7]
y = [99, 86, 87, 88]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
#barchart
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()

```

```python
#Histogram
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data, bins=4)
plt.title("Histogram")
plt.show()
```

```python
#PieChart
labels = ['A', 'B', 'C']
sizes = [50, 30, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
#Boxplot
data = [7, 15, 13, 18, 22, 25, 30]
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

```python
#LinePlot
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x=[1, 2, 3, 4], y=[10, 20, 25, 30])
plt.title("Line Plot - Seaborn")
plt.show()
```

```python
#ScatterPlot
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 20, 25, 30],
    'group': ['A', 'A', 'B', 'B']
})
sns.scatterplot(data=df, x='x', y='y', hue='group')
plt.title("Scatter Plot - Seaborn")
plt.show()

```

```python
#Barchart
df = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [10, 20, 15]
})
sns.barplot(x='Category', y='Value', data=df)
plt.title("Bar Chart - Seaborn")
plt.show()
```

```python
#Histogram
data = [1,2,2,3,3,3,4,4,4,4]
sns.histplot(data, bins=4, kde=True)
plt.title("Histogram - Seaborn")
plt.show()
```

```python
#Boxplot
df = pd.DataFrame({
    'value': [7, 15, 13, 18, 22, 25, 30],
    'group': ['A', 'A', 'A', 'B', 'B', 'B', 'B']
})
sns.boxplot(x='group', y='value', data=df)
plt.title("Box Plot - Seaborn")
plt.show()
```

```python
#Heatmap
import numpy as np

data = np.array([[1, 2], [3, 4]])
sns.heatmap(data, annot=True)
plt.title("Heatmap")
plt.show()
```
